### Проверка эффективности обогащения

Проверим, что эффективность обогащения рассчитана правильно. Вычислим её на обучающей выборке для признака rougher.output.recovery. Найдем MAE между нашими расчётами и значением признака.

Чтобы понять какие столбцы нам нужны для проверки посмотрим на них еще раз (а также на сам столбец rougher.output.recovery):

```python
df_train['rougher.output.recovery'].head()
```

Видим, что нужный нам столбец представлен в процентах.

```python
df_train.columns
```

Мы видим следующие группы стобцов:
- date - момент проведение этапа;
- Признаки, относящиеся к разным стадиям процееса (первая часть имени столбца): финальной версии вещества после всех обработок (final), после флотации (rougher), и разных уровней очистки (primary_cleaner, secondary_cleaner);
- Внутри каждого этапа мы видим возможные подэтапы в процессе: output, input, calculation, state;
- А внутри каждого поэтапа мы видим разные параметры разных веществ, соответсвующие этому подэтапу. Среди них есть и понятные и нужные нам для расчетов: к примеру, с окончанием _au. Так и неизвстные: к примеру, с окончанием _sol;
- Целевой признаки 'rougher.output.recovery и final.output.recovery'.

Для вычисления нам нужны:
- C — доля золота в концентрате после флотации;
- F — доля золота в сырье до флотации;
- T — доля золота в отвальных хвостах после флотации.

Чтобы использовать это в формуле: Recovery = (100 * C * (F-T)) / (F * (C-T)).

Итак:
- C - это столбец 'rougher.output.concentrate_au',
- F - это столбец 'rougher.input.feed_au',
- T - это столбец 'rougher.output.tail_au'.

Создадим наш новый расчетный столбец:

```python
df_train['rougher.output.recovery_new'] = (
    100 * df_train['rougher.output.concentrate_au'] * (
        df_train['rougher.input.feed_au'] - df_train['rougher.output.tail_au'])
) / (
    df_train['rougher.input.feed_au'] * (
        df_train['rougher.output.concentrate_au'] - df_train['rougher.output.tail_au'])
)
```

```python
df_train['rougher.output.recovery_new'].head()
```

Видим, что порядок цифр совпал. Теперь посчитаем MAE (среднее абсолютное отклонение) между нашими расчётами и значением признака.

```python
mean_absolute_error(df_train['rougher.output.recovery_new'], df_train['rougher.output.recovery'])
```

Видим, что это значение очень маленькое: 1.1455109954099771e-14. Это хорошо, значит вычисления в исходных данных верные.

Чтобы решить эти диллемы с похожестью данных не на глаз, проверим гипотезу:
- среднее значение в признаках с размером в тестовой и обучающей выборках совпадает.

Это нулевая гипотеза H₀: средние у совокупностей равны.

Чтобы проверить гипотезу о равенстве среднего двух генеральных совокупностей по взятым из них выборкам, применим метод scipy.stats.ttest_ind (array1, array2, equal_var). Он как раз поможет нам понять зависимость.

Пороговое значение alpha зададим как 0.05 (критический уровень статистической значимости, если p-value окажется меньше него - отвергнем гипотезу). 

Единственное необходимо понять насколько различаются дисперсии в выборках. Для этого применим тест Левене.

```python
alpha = 0.05

sample_1 = df_train['rougher.input.feed_size']
sample_2 = df_test['rougher.input.feed_size']
```

```python
levene_result = st.levene(sample_1, sample_2)
levene_result
```

```python
def equal_var_result(pvalue):
    if pvalue < alpha:
        return 'Берем equal_var=False'
    else:
        return 'Берем equal_var=True'

equal_var_result(levene_result.pvalue)
```

P-значение получилось больше, чем порог, однозначно equal_var=True. Приступим к t тесту.

```python
results = st.ttest_ind(sample_1, sample_2, equal_var=True)
results.pvalue
```

```python
def hypothesis_result(pvalue):
    if pvalue < alpha:
        return 'Отвергаем нулевую гипотезу'
    else:
        return 'Не получилось отвергнуть нулевую гипотезу'

hypothesis_result(results.pvalue)
```

Отвергаем нулевую гипотезу, а значит и считаем распределения различными. Попробуем убрать длинные и редкие хвосты в обоих выборках.

### Функция для sMape

Определим функцию для расчета эффективности модели. Она должна принимать на вход: 2 набора с предсказаниями, 2 набора с таргетами.

Создадим сначала подфункцию.

```python
def sMape(predictions, target):
    result = 0
    result = sum(abs(predictions - target) / ((abs(predictions) + abs(target)) * 0.5))
    result = result * 100 * (1/len(target))                                            
    return result
```

```python
def final_sMape(sMape1, sMape2):
    return sMape1*0.25 + sMape2*0.75
```
